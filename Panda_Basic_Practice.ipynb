```python
# prompt: can u create a histogram on df on sat column can you change the colour to green and edges to black

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('Universities.csv')

plt.hist(df['SAT'], color='green', edgecolor='black')
plt.xlabel('SAT Scores')
plt.ylabel('Frequency')
plt.title('Histogram of SAT Scores')
plt.show()
```

```python
plt.plot(df['SAT'],df['Accept'],'rs')
plt.xlabel('SAT Scores')
plt.ylabel('Accept')
plt.title('Scatter Plot of SAT Scores vs Accept')
plt.show()
```
